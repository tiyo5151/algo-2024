```python
import matplotlib.pyplot as plt
import math
```

```python
xs = []
ys = []
for x in range(-10, 11):
    y = x**2
    xs.append(x)
    ys.append(y)
plt.plot(xs, ys, label="y = x * x")
plt.title("Quadratic curve")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
xs = []
ys_sin = []
ys_cos = []
for i in range(0, 100):
    x = i * 0.1
    xs.append(x)
    ys_sin.append(math.sin(x))
    ys_cos.append(math.cos(x))
plt.plot(xs, ys_sin, label="y = sin(x)")
plt.plot(xs, ys_cos, label="y = cos(x)")
plt.title("Trigonometric function")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
xs = []
ys = []
for i in range(10, 101):
    x = i * 0.1
    xs.append(x)
    ys.append(x)
plt.plot(xs, ys, label="$y = x$")
plt.xlim([1, 10])
plt.title("Graph of $y = x$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
xs = []
ys = []
for i in range(10, 101):
    x = i * 0.1
    xs.append(x)
    ys.append(2**x)
plt.plot(xs, ys, label="$y = 2^x$")
plt.xlim([1, 10])
plt.title("Graph of $y = 2^x$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
xs = []
ys = []
for i in range(10, 101):
    x = i * 0.1
    xs.append(x)
    ys.append(math.log(x))
plt.plot(xs, ys, label=" $y = \log x$")
plt.xlim([1, 10])
plt.title("Graph of $y = \log x$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
