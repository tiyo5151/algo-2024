# 情報理論入門

## ビット(bit)

- 0,1 の列で表される
- ビット数は情報量の単位としての芋を持つ
  - ex. 5bit の情報 -> $2^{5}$=32 通りの情報
- **1 バイト(B)=8[bit]**

  ### 単位

  - T(Tera) = $10^{12}$
  - G(Giga) = $10^{9}$
  - M(Mega) = $10^{6}$
  - K(Kilo) = $10^{3}$
  - c(centi) = $10^{-2}$
  - m(milli) = $10^{-3}$
  - μ(micro) = $10^{-6}$
  - n(nano) = $10^{-9}$
  - p(pico) = $10^{-12}$

- 情報量は加法性が成り立つ

## 情報量の定義

確率 p で発生する事象の情報量 I
$$I = -\log_2 p \;\text{[bit]}$$

ex. A さんのクラスがクラス 2 の時の事象の情報量(クラスは 20 あるとする)
$$I = -\log_2 \frac{1}{20} \approx 4.3219 \;\text{[bit]}$$

- bit と同様加法性がある

## 符号化

情報をデジタル情報としてビット列(0,1)にする処理のこと

## 情報エントロピー

情報をできる限り少ないビット数で表現したい
-> 情報量の平均のようなものが欲しい
↑ のために情報エントロピーの概念を導入

## 情報エントロピーの定義

情報エントロピー H は情報量の期待値として定義される
$$H = E[I] = -\sum_{i=1}^n p_i \log_2 p_i \;\text{[bit]}$$
