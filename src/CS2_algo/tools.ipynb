# 情報量計算ツール


```python
import math


def calc_info():
    result = []
    multi_result = 1
    while True:
        try:
            a = input("数字を入力してください(終了時はそのままenter) ")
            if a == "":
                break

            # Handle fractions in the input
            if "/" in a:
                num, denom = map(float, a.split("/"))
                a = num / denom
            else:
                a = float(a)

            print(a)
            result.append(a)
        except (ValueError, ZeroDivisionError):
            print("正しい数値または分数を入力してください")
            continue

    for i in result:
        multi_result *= i

    return -math.log2(multi_result)


calc_info()
```

# 情報エントロピー計算ツール


```python
def calc_entropy():
    result = []
    while True:
        try:
            a = input("数字を入力してください(終了時はそのままenter) ")
            if a == "":
                break
            if "/" in a:
                num, denom = map(float, a.split("/"))
                a = num / denom
            else:
                a = float(a)
            print(a)
            result.append(a)
        except (ValueError, ZeroDivisionError):
            print("正しい数値または分数を入力してください")
            continue

    # 確率の合計が1になるかチェック
    if abs(sum(result) - 1.0) > 1e-10:
        return "エラー：確率の合計が1になっていません"

    # ループの外でfinal_resultを初期化
    final_result = 0
    for i in result:
        try:
            # エントロピーの計算：-p * log2(p)
            final_result += -i * math.log2(i)
        except ValueError:
            # 0 * log(0) = 0 として扱う
            continue

    return final_result


calc_entropy()
```

# 最大公約数を求めるツール(一個目がユークリッド、二個目がなし)


```python
def gcd():
    try:
        a = int(input("最初の数字を入力してください："))
        b = int(input("次の数字を入力してください："))
    except ValueError:
        print("エラー：整数を入力してください")
        return

    if a < b:
        tmp = a
        a = b
        b = tmp
    while True:
        r = a % b
        if r == 0:
            return b
        a = b
        b = r


gcd()
```

```python
def get_prime_factors(n):
    """数値を素因数分解し、素因数とその指数を辞書で返す"""
    factors = {}
    i = 2
    while i * i <= n:
        if n % i == 0:
            count = 0
            while n % i == 0:
                count += 1
                n //= i
            factors[i] = count
        i += 1
    if n > 1:
        factors[n] = 1
    return factors


def calculate_gcd(a, b):
    """2つの数の最大公約数を計算する"""
    if a == 0 or b == 0:
        return max(abs(a), abs(b))

    # 負の数を正の数に変換
    a, b = abs(a), abs(b)

    # 両方の数を素因数分解
    factors_a = get_prime_factors(a)
    factors_b = get_prime_factors(b)

    # 共通する素因数の中で、より小さい指数を選択
    gcd = 1
    for prime in set(factors_a.keys()) & set(factors_b.keys()):
        gcd *= prime ** min(factors_a[prime], factors_b[prime])

    return gcd


def get_valid_integer_input(prompt):
    """ユーザーから有効な整数を入力として受け取る"""
    while True:
        try:
            value = int(input(prompt))
            return value
        except ValueError:
            print("エラー: 有効な整数を入力してください。")


def main():
    print("最大公約数を計算します。")
    print("2つの整数を入力してください。")

    # 1つ目の数の入力
    num1 = get_valid_integer_input("1つ目の数を入力: ")

    # 2つ目の数の入力
    num2 = get_valid_integer_input("2つ目の数を入力: ")

    # 最大公約数の計算と表示
    result = calculate_gcd(num1, num2)
    print(f"\n{num1}と{num2}の最大公約数は: {result}")


if __name__ == "__main__":
    main()
```

# 二分探索比較回数確認ツール


```python
def binary_search():
    lst = sorted(input("リストを入力してください(スペースで区切る)：").split())
    x = input("探す値を入力してください：")
    a = 0
    start = 0
    end = len(lst)
    while start < end:
        a += 1
        center = start + (end - start) // 2
        # print(lst [center])
        if lst[center] == x:
            return True, a
        elif lst[center] > x:
            end = center
        else:
            start = center + 1
    return False, a


binary_search()
```

## $logn$


```python
import time
import random


def binary_search(arr, target):
    """O(log n)の計算量を持つ二分探索"""
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1


def measure_execution_time_logn(n, num_searches=1000):
    """要素数nの配列で二分探索の実行時間を測定"""
    # ソート済みの配列を生成
    arr = sorted(range(n))

    # 実行時間の測定開始
    start_time = time.time()

    # 複数回の検索を実行して平均的な実行時間を得る
    for _ in range(num_searches):
        target = random.randint(0, n - 1)
        binary_search(arr, target)

    # 実行時間の測定終了
    end_time = time.time()

    # 平均実行時間を返す
    return (end_time - start_time) / num_searches


def main():
    print("O(log n)の計算量の実行時間比較\n")

    # ユーザーから入力を受け取る
    while True:
        try:
            n1 = int(input("1つ目の要素数(n1)を入力してください: "))
            n2 = int(input("2つ目の要素数(n2)を入力してください: "))
            if n1 <= 0 or n2 <= 0:
                print("正の整数を入力してください。")
                continue
            break
        except ValueError:
            print("有効な整数を入力してください。")

    # それぞれの要素数での実行時間を測定
    time1 = measure_execution_time_logn(n1)
    time2 = measure_execution_time_logn(n2)

    # 実行時間の比率を計算
    time_ratio = time2 / time1

    # 理論上の計算量の比率を計算（O(log n)の場合）
    import math

    theoretical_ratio = math.log2(n2) / math.log2(n1)

    # 結果の表示
    print(f"\n結果:")
    print(f"n1 = {n1}の実行時間: {time1:.8f}秒")
    print(f"n2 = {n2}の実行時間: {time2:.8f}秒")
    print(f"実際の実行時間の比率: {time_ratio:.2f}倍")
    print(f"理論上の計算量の比率: {theoretical_ratio:.2f}倍")

    # 要素数の増加に対する影響の説明
    ratio_n = n2 / n1
    print(f"\n参考情報:")
    print(f"要素数の増加率: {ratio_n:.2f}倍")
    print(f"もしO(n)の場合の理論上の比率: {ratio_n:.2f}倍")
    print(f"もしO(n²)の場合の理論上の比率: {ratio_n**2:.2f}倍")
    print(f"実際のO(log n)の比率: {theoretical_ratio:.2f}倍")
    print(f"\n注: 実行時間は環境やその他の要因によって変動する可能性があります。")


if __name__ == "__main__":
    main()
```

```python
import time
import random


def linear_search(arr, target):
    """O(n)の計算量を持つ線形探索"""
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1


def measure_execution_time_linear(n, num_searches=100):
    """要素数nの配列で線形探索の実行時間を測定"""
    # 配列を生成
    arr = list(range(n))

    # 実行時間の測定開始
    start_time = time.time()

    # 複数回の検索を実行して平均的な実行時間を得る
    for _ in range(num_searches):
        target = random.randint(0, n - 1)
        linear_search(arr, target)

    # 実行時間の測定終了
    end_time = time.time()

    # 平均実行時間を返す
    return (end_time - start_time) / num_searches


def format_time(seconds):
    """時間を読みやすい形式にフォーマット"""
    if seconds < 0.000001:  # 1マイクロ秒未満
        return f"{seconds * 1000000000:.2f} ナノ秒"
    elif seconds < 0.001:  # 1ミリ秒未満
        return f"{seconds * 1000000:.2f} マイクロ秒"
    elif seconds < 1:  # 1秒未満
        return f"{seconds * 1000:.2f} ミリ秒"
    else:
        return f"{seconds:.2f} 秒"


def main():
    print("O(n)の計算量の実行時間比較\n")

    # ユーザーから入力を受け取る
    while True:
        try:
            n1 = int(input("1つ目の要素数(n1)を入力してください: "))
            n2 = int(input("2つ目の要素数(n2)を入力してください: "))
            if n1 <= 0 or n2 <= 0:
                print("正の整数を入力してください。")
                continue
            break
        except ValueError:
            print("有効な整数を入力してください。")

    # それぞれの要素数での実行時間を測定
    time1 = measure_execution_time_linear(n1)
    time2 = measure_execution_time_linear(n2)

    # 実行時間の比率を計算
    time_ratio = time2 / time1

    # 理論上の計算量の比率を計算（O(n)の場合）
    theoretical_ratio = n2 / n1

    # 結果の表示
    print(f"\n結果:")
    print(f"n1 = {n1}の実行時間: {format_time(time1)}")
    print(f"n2 = {n2}の実行時間: {format_time(time2)}")
    print(f"実際の実行時間の比率: {time_ratio:.2f}倍")
    print(f"理論上の計算量の比率: {theoretical_ratio:.2f}倍")

    # 要素数の増加に対する影響の説明
    import math

    print(f"\n参考情報:")
    print(f"要素数の増加率: {theoretical_ratio:.2f}倍")
    print(f"O(log n)の場合の理論上の比率: {math.log2(n2)/math.log2(n1):.2f}倍")
    print(f"O(n)の場合の理論上の比率: {theoretical_ratio:.2f}倍")
    print(f"O(n²)の場合の理論上の比率: {(n2/n1)**2:.2f}倍")

    print(f"\n注:")
    print("1. 実行時間は環境やその他の要因によって変動する可能性があります。")
    print("2. 小さすぎる要素数では、システムのオーバーヘッドにより")
    print("   理論値との誤差が大きくなる可能性があります。")


if __name__ == "__main__":
    main()
```

## $nlogn$


```python
import time
import random
import math


def merge_sort(arr):
    """O(n log n)の計算量を持つマージソート"""
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    # 再帰的に分割と統合
    left = merge_sort(left)
    right = merge_sort(right)

    return merge(left, right)


def merge(left, right):
    """2つのソート済み配列をマージ"""
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


def measure_execution_time_nlogn(n, num_trials=5):
    """要素数nの配列でマージソートの実行時間を測定"""
    total_time = 0

    for _ in range(num_trials):
        # ランダムな配列を生成
        arr = [random.randint(1, 1000000) for _ in range(n)]

        # 実行時間の測定開始
        start_time = time.time()

        # マージソートの実行
        merge_sort(arr.copy())  # 元の配列を変更しないようにcopy

        # 実行時間の測定終了
        end_time = time.time()
        total_time += end_time - start_time

    # 平均実行時間を返す
    return total_time / num_trials


def format_time(seconds):
    """時間を読みやすい形式にフォーマット"""
    if seconds < 0.000001:  # 1マイクロ秒未満
        return f"{seconds * 1000000000:.2f} ナノ秒"
    elif seconds < 0.001:  # 1ミリ秒未満
        return f"{seconds * 1000000:.2f} マイクロ秒"
    elif seconds < 1:  # 1秒未満
        return f"{seconds * 1000:.2f} ミリ秒"
    else:
        return f"{seconds:.2f} 秒"


def calculate_theoretical_ratio(n1, n2, complexity_type):
    """理論上の計算量の比率を計算"""
    if complexity_type == "O(log n)":
        return math.log2(n2) / math.log2(n1)
    elif complexity_type == "O(n)":
        return n2 / n1
    elif complexity_type == "O(n log n)":
        return (n2 * math.log2(n2)) / (n1 * math.log2(n1))
    elif complexity_type == "O(n²)":
        return (n2 * n2) / (n1 * n1)
    return 0


def main():
    print("O(n log n)の計算量の実行時間比較\n")

    # ユーザーから入力を受け取る
    while True:
        try:
            n1 = int(input("1つ目の要素数(n1)を入力してください: "))
            n2 = int(input("2つ目の要素数(n2)を入力してください: "))
            if n1 <= 0 or n2 <= 0:
                print("正の整数を入力してください。")
                continue
            break
        except ValueError:
            print("有効な整数を入力してください。")

    # それぞれの要素数での実行時間を測定
    print("\n計測中...\n")
    time1 = measure_execution_time_nlogn(n1)
    time2 = measure_execution_time_nlogn(n2)

    # 実行時間の比率を計算
    time_ratio = time2 / time1

    # 理論上の計算量の比率を計算
    theoretical_ratio = calculate_theoretical_ratio(n1, n2, "O(n log n)")

    # 結果の表示
    print(f"結果:")
    print(f"n1 = {n1}の実行時間: {format_time(time1)}")
    print(f"n2 = {n2}の実行時間: {format_time(time2)}")
    print(f"実際の実行時間の比率: {time_ratio:.2f}倍")
    print(f"理論上の計算量の比率: {theoretical_ratio:.2f}倍")

    # 他の計算量との比較
    print(f"\n他の計算量との比較:")
    print(f"要素数の増加率: {n2/n1:.2f}倍")
    print(f"O(log n)の場合: {calculate_theoretical_ratio(n1, n2, 'O(log n)'):.2f}倍")
    print(f"O(n)の場合: {calculate_theoretical_ratio(n1, n2, 'O(n)'):.2f}倍")
    print(f"O(n log n)の場合: {theoretical_ratio:.2f}倍")
    print(f"O(n²)の場合: {calculate_theoretical_ratio(n1, n2, 'O(n²)'):.2f}倍")

    print(f"\n注:")
    print("1. 実行時間は環境やその他の要因によって変動する可能性があります。")
    print(
        "2. マージソートは安定ソートであり、最悪の場合でもO(n log n)の計算量を保証します。"
    )
    print("3. メモリ使用量はO(n)です。")


if __name__ == "__main__":
    main()
```

## $n^C$


```python
import time
import random


def bubble_sort(arr):
    """O(n²)の計算量を持つバブルソート"""
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr


def measure_execution_time(n):
    """要素数nの配列をソートする実行時間を測定"""
    # ランダムな要素を持つ配列を生成
    arr = [random.randint(1, 10000) for _ in range(n)]

    # 実行時間の測定開始
    start_time = time.time()

    # アルゴリズムの実行
    bubble_sort(arr.copy())  # 元の配列を変更しないようにcopyを使用

    # 実行時間の測定終了
    end_time = time.time()

    return end_time - start_time


def main():
    # ユーザーから入力を受け取る
    while True:
        try:
            n1 = int(input("1つ目の要素数(n1)を入力してください: "))
            n2 = int(input("2つ目の要素数(n2)を入力してください: "))
            if n1 <= 0 or n2 <= 0:
                print("正の整数を入力してください。")
                continue
            break
        except ValueError:
            print("有効な整数を入力してください。")

    # それぞれの要素数での実行時間を測定
    time1 = measure_execution_time(n1)
    time2 = measure_execution_time(n2)

    # 実行時間の比率を計算
    time_ratio = time2 / time1

    # 理論上の計算量の比率を計算（O(n²)の場合）
    theoretical_ratio = (n2 * n2) / (n1 * n1)

    # 結果の表示
    print(f"\n結果:")
    print(f"n1 = {n1}の実行時間: {time1:.6f}秒")
    print(f"n2 = {n2}の実行時間: {time2:.6f}秒")
    print(f"実際の実行時間の比率: {time_ratio:.2f}倍")
    print(f"理論上の計算量の比率: {theoretical_ratio:.2f}倍")
    print(f"\n注: 実行時間は環境やその他の要因によって変動する可能性があります。")


if __name__ == "__main__":
    main()
```

## $C^n$


```python
import time
import math


def fibonacci_recursive(n):
    """O(2^n)の計算量を持つフィボナッチ数列の計算（再帰的実装）"""
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)


def measure_execution_time_exponential(n, num_trials=3):
    """要素数nでのフィボナッチ数列計算の実行時間を測定"""
    total_time = 0

    for _ in range(num_trials):
        # 実行時間の測定開始
        start_time = time.time()

        # フィボナッチ数列の計算
        fibonacci_recursive(n)

        # 実行時間の測定終了
        end_time = time.time()
        total_time += end_time - start_time

    # 平均実行時間を返す
    return total_time / num_trials


def format_time(seconds):
    """時間を読みやすい形式にフォーマット"""
    if seconds < 0.000001:  # 1マイクロ秒未満
        return f"{seconds * 1000000000:.2f} ナノ秒"
    elif seconds < 0.001:  # 1ミリ秒未満
        return f"{seconds * 1000000:.2f} マイクロ秒"
    elif seconds < 1:  # 1秒未満
        return f"{seconds * 1000:.2f} ミリ秒"
    else:
        return f"{seconds:.2f} 秒"


def calculate_theoretical_ratio(n1, n2, complexity_type):
    """理論上の計算量の比率を計算"""
    if complexity_type == "O(log n)":
        return math.log2(n2) / math.log2(n1)
    elif complexity_type == "O(n)":
        return n2 / n1
    elif complexity_type == "O(n log n)":
        return (n2 * math.log2(n2)) / (n1 * math.log2(n1))
    elif complexity_type == "O(n²)":
        return (n2 * n2) / (n1 * n1)
    elif complexity_type == "O(2^n)":
        return 2 ** (n2 - n1)
    return 0


def main():
    print("O(2^n)の計算量の実行時間比較\n")
    print("警告: 指数時間アルゴリズムは入力値が大きいと非常に時間がかかります。")
    print("推奨される入力値の範囲: 20-35\n")

    # ユーザーから入力を受け取る
    while True:
        try:
            n1 = int(input("1つ目の値(n1)を入力してください: "))
            n2 = int(input("2つ目の値(n2)を入力してください: "))
            if n1 <= 0 or n2 <= 0:
                print("正の整数を入力してください。")
                continue
            if n1 > 35 or n2 > 35:
                print("警告: 大きすぎる値は計算に非常に時間がかかる可能性があります。")
                confirm = input("続行しますか？(y/n): ")
                if confirm.lower() != "y":
                    continue
            break
        except ValueError:
            print("有効な整数を入力してください。")

    # それぞれの値での実行時間を測定
    print("\n計測中...\n")
    time1 = measure_execution_time_exponential(n1)
    time2 = measure_execution_time_exponential(n2)

    # 実行時間の比率を計算
    time_ratio = time2 / time1

    # 理論上の計算量の比率を計算
    theoretical_ratio = calculate_theoretical_ratio(n1, n2, "O(2^n)")

    # 結果の表示
    print(f"結果:")
    print(f"n1 = {n1}の実行時間: {format_time(time1)}")
    print(f"n2 = {n2}の実行時間: {format_time(time2)}")
    print(f"実際の実行時間の比率: {time_ratio:.2f}倍")
    print(f"理論上の計算量の比率: {theoretical_ratio:.2f}倍")

    # 他の計算量との比較
    print(f"\n他の計算量との比較（n2/n1の増加に対する理論上の増加率）:")
    print(f"O(log n)の場合: {calculate_theoretical_ratio(n1, n2, 'O(log n)'):.2f}倍")
    print(f"O(n)の場合: {calculate_theoretical_ratio(n1, n2, 'O(n)'):.2f}倍")
    print(
        f"O(n log n)の場合: {calculate_theoretical_ratio(n1, n2, 'O(n log n)'):.2f}倍"
    )
    print(f"O(n²)の場合: {calculate_theoretical_ratio(n1, n2, 'O(n²)'):.2f}倍")
    print(f"O(2^n)の場合: {theoretical_ratio:.2f}倍")

    print(f"\n注:")
    print("1. 実行時間は環境やその他の要因によって変動する可能性があります。")
    print("2. 指数時間アルゴリズムは入力サイズが少し大きくなるだけで")
    print("   実行時間が劇的に増加します。")
    print("3. この実装は教育目的であり、実際の応用では効率的な")
    print("   動的計画法などの手法を使用すべきです。")


if __name__ == "__main__":
    main()
```

# 中置記法から前置記法への変換プラン

1. スタックを使用して演算子と被演算子を管理
2. 式を右から左に読む
3. 括弧の処理を含める
4. 優先順位に基づいて演算子を配置


```python
def infix_to_prefix(expression):
    # 演算子の優先順位
    precedence = {"+": 1, "-": 1, "*": 2, "/": 2, "^": 3}

    # 式を反転して処理（括弧も反転）
    expression = expression.replace("(", "#")
    expression = expression.replace(")", "(")
    expression = expression.replace("#", ")")
    tokens = expression.split()
    tokens.reverse()

    output = []
    operators = []

    for token in tokens:
        if token.isdigit() or (token[0] == "-" and token[1:].isdigit()):
            output.append(token)
        elif token in precedence:
            while (
                operators
                and operators[-1] != ")"
                and precedence.get(operators[-1], 0) > precedence[token]
            ):
                output.append(operators.pop())
            operators.append(token)
        elif token == ")":
            operators.append(token)
        elif token == "(":
            while operators and operators[-1] != ")":
                output.append(operators.pop())
            if operators and operators[-1] == ")":
                operators.pop()

    while operators:
        if operators[-1] == ")":
            return "Error: 括弧の不一致"
        output.append(operators.pop())

    # 結果を反転して前置記法に
    output.reverse()
    return " ".join(output)


# 使用例
expression = "3 + 4 * 2 / ( 1 - 5 )"
prefix = infix_to_prefix(expression)
print(f"中置記法: {expression}")
print(f"前置記法: {prefix}")
```

この実装により、例えば:

- "3 + 4" → "+ 3 4"
- "1 + 2 _ 3" → "+ 1 _ 2 3"
  のように変換されます。


# 中置記法から後置記法（RPN）への変換プラン

1. 変換ステップ:
   - スタックを用いて演算子を管理
   - 優先順位に基づいて処理
   - 左から右へ式を読む
   - 括弧の処理を実装


```python
def infix_to_postfix(expression):
    # 演算子の優先順位
    precedence = {"+": 1, "-": 1, "*": 2, "/": 2, "^": 3}

    output = []  # 出力用リスト
    operators = []  # 演算子スタック

    # スペースで区切られた式を処理
    tokens = expression.split()

    for token in tokens:
        # 数値の場合
        if token.isdigit() or (token[0] == "-" and token[1:].isdigit()):
            output.append(token)

        # 演算子の場合
        elif token in precedence:
            while (
                operators
                and operators[-1] != "("
                and precedence.get(operators[-1], 0) >= precedence[token]
            ):
                output.append(operators.pop())
            operators.append(token)

        # 開き括弧
        elif token == "(":
            operators.append(token)

        # 閉じ括弧
        elif token == ")":
            while operators and operators[-1] != "(":
                output.append(operators.pop())
            if operators and operators[-1] == "(":
                operators.pop()

    # スタックに残った演算子を出力に追加
    while operators:
        if operators[-1] == "(":
            return "Error: 括弧の不一致"
        output.append(operators.pop())

    return " ".join(output)


# 使用例
expression = "3 + 4 * 2 / ( 1 - 5 )"
postfix = infix_to_postfix(expression)
print(f"中置記法: {expression}")
print(f"後置記法: {postfix}")
```
